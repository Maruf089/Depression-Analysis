```

class My_LinearRegression:
    def __init__(self, max_iteration = 1000, max_MSE = None, learning_rate = 0.001):
        self.max_iteration=max_iteration
        self.learning_rate = learning_rate
        self.max_MSE = max_MSE

    def fit_function(self, X, Y):
       
        if isinstance(X,pd.Series):
            X = X.to_frame()
        Y = Y.to_frame()
        self.n = len(X)
        self.coeff = [0 for _ in range(len(X.columns))]
        self.intercept = 0
        self.MSE=[]
        self.n_iteration = 0
        X = X.values
        Y = Y.values

        while(True):
            y_pred = np.sum(X*self.coeff,axis=1) + self.intercept
            y_pred = y_pred.reshape(self.n,1)

            current_MSE = np.square(np.subtract(Y,y_pred)).mean()
            self.MSE.append(current_MSE)
            
            ## compute gradients
            df_dm = -2*(X*(Y-y_pred).reshape(self.n,1)).mean(axis=0)
            df_dc = -2*(Y-y_pred).mean()
            
            ##update parameters
            self.coeff = self.coeff - df_dm * self.learning_rate
            self.intercept = self.intercept - df_dc * self.learning_rate

            self.n_iteration = self.n_iteration + 1
            if self.n_iteration >= self.max_iteration:
                break
  

    def prediction_function(self,X):
        if isinstance(X,list):
            X = [X]
        if isinstance(X,pd.Series):
            X = X.to_frame()
        if isinstance(X,pd.DataFrame):
            X = X.values
        n = len(X)
        return (np.sum(X*self.coeff,axis=1) + self.intercept).reshape(n,1)
```

```

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.metrics import mean_squared_error
```

```
df = pd.read_csv("/content/drive/MyDrive/Dataset/Depression Analysis/Depression and Happiness Factor Analysis , Maruf's Copy (Responses) - Form Responses 1.csv")


Scale = df["On a scale of 1-100, how would you express this feeling?"]
df.drop(["label","On a scale of 1-100, how would you express this feeling?","Timestamp"],axis=1,inplace=True)

df_Encoded = pd.get_dummies(df)

x_train, x_test, y_train, y_test = train_test_split(df_Encoded, Scale, test_size= 0.2, random_state=1)
```

```
model = My_LinearRegression()

model.fit_function(x_train, y_train)

y_pred = model.prediction_function(x_train)

r2 = r2_score(y_train, y_pred)

print("Training R2 : {}".format(r2))
train_mse = mean_squared_error(y_train, y_pred)

print("Training mse : {}".format(train_mse))

y_pred = model.prediction_function(x_test)
r2 = r2_score(y_test, y_pred)

print("Testing r2 : {}".format(r2))

test_mse = mean_squared_error(y_test, y_pred)
print("Testing mse : {}".format(test_mse))


```
